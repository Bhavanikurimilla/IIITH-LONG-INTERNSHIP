## **CDF**

```python
class_1=haberman.loc[haberman["status"]==1]
class_2=haberman.loc[haberman["status"]==2]
```

```python
counts,bin_edges=np.histogram(class_1['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)


cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],cdf,label='cdf')
plt.xlabel('AGE')
plt.legend()
plt.show()
```
